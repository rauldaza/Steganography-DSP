```python
# Initialize an empty list to store the Fourier transform of each slice
def fft_window(length, width, fs, senal):
    samples_per_slice = int((length/width*len(senal))**(0.5))
    #samples_per_slice = int(fs * slice_duration)
    
    for i in range(0, len(senal), samples_per_slice):
        # Extract a slice of audio
        audio_slice = senal[i:i + samples_per_slice]

        # Compute the Fourier transform of the slice
        slice_fft = fft(audio_slice)
        # Append the slice's Fourier transform to the list
        if (i==0):
            audio_fft_slices = slice_fft.reshape(slice_fft.size,1)
        else:
            if (slice_fft.size != len(audio_fft_slices)):
                break
            audio_fft_slices = np.hstack((audio_fft_slices,slice_fft.reshape((slice_fft.size,1))))
    return audio_fft_slices
```

```python
# Initialize an empty list to store the magnitudes of the Fourier transforms
def fft_window_split(audio_fft_slices):
    # Initialize empty lists to store magnitudes and phases
    magnitude_slices = []
    phase_slices = []

    for slice_fft in audio_fft_slices:
        # Compute the magnitude and phase of the slice's Fourier transform
        magnitude_slice = np.abs(slice_fft)
        phase_slice = np.angle(slice_fft)

        # Append the magnitude and phase slices to their respective lists
        magnitude_slices.append(magnitude_slice)
        phase_slices.append(phase_slice)
    return np.array(magnitude_slices), np.array(phase_slices)
```

```python
def reconstruct_fft_split(magnitude_slices,phase_slices):
    # Initialize an empty array to store the reconstructed audio
    reconstructed_audio = np.array([])
    
    for mag_slice, phase_slice in zip(magnitude_slices.T, phase_slices.T):
        # Combine the magnitude and phase to create the complex spectrum
        complex_spectrum = mag_slice * np.exp(1j * phase_slice)

        # Compute the inverse Fourier transform of the complex spectrum
        slice_ifft = ifft(complex_spectrum)

        # Append the slice's inverse Fourier transform to the result array
        reconstructed_audio = np.append(reconstructed_audio, slice_ifft.real)
    return reconstructed_audio
```

```python
def reconstruc_fft(audio_fft_slices):
    # Initialize an empty array to store the inverse Fourier transform of each slice
    reconstructed_audio = np.array([])
    for slice_fft in audio_fft_slices:
        # Compute the inverse Fourier transform of the slice
        slice_ifft = ifft(slice_fft)
        
        # Append the slice's inverse Fourier transform to the result array
        reconstructed_audio = np.append(reconstructed_audio, slice_ifft.real)
    return reconstructed_audio
```

```python
window = fft_window(*(im.shape)[0:2],fs,senal)
mag, phase = fft_window_split(window)
mag_norm = mag/np.max(mag)*0.1
plt.imshow(mag_norm)
plt.colorbar()
```

```python
phase_norm = (phase/(2*np.pi)+0.5)*0.1
plt.imshow(phase_norm)
plt.colorbar()
```
